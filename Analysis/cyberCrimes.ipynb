```python
import pandas as pd
data = pd.read_csv("LossFromNetCrime.csv")
data.head()
```

```python
# Function to summarize dataset information
def data_summary(df):
    print("Dataset Shape:", df.shape)
    print("\nDataset Info:")
    print(df.info())
    print("\nDataset Description:")
    print(df.describe(include='all'))
    print("\nMissing Values:")
    print(df.isnull().sum())

# Call the function
data_summary(data)
```

```python
import matplotlib.pyplot as plt
def plot_complaints_growth(df):
    complaints_columns = [col for col in df.columns if 'Complaints' in col]
    df_complaints = df[['Country'] + complaints_columns].set_index('Country')
    df_complaints.T.plot(figsize=(12, 6), marker='o')
    plt.title('Complaints Growth Over Years by Country', fontsize=14)
    plt.ylabel('Number of Complaints', fontsize=12)
    plt.xlabel('Year', fontsize=12)
    plt.xticks(rotation=45)
    plt.legend(title='Country')
    plt.grid()
    plt.show()

# Call the function
plot_complaints_growth(data)
```
